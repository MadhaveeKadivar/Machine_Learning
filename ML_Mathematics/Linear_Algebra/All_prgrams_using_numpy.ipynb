1.Write a python program to add below matrices
X = [[12,7,3],
[4 ,5,6],
[7 ,8,9]]
Y = [[5,8,1],
[6,7,3],
[4,5,9]]

```python
import numpy as np
x = np.array([[12,7,3],[4,5,6],[7,8,9]])
y = np.array([[5,8,1],[6,7,3],[4,5,9]])
result = np.add(x, y)
print("Addition of matrices x and y is : \n")
print(result)
```
