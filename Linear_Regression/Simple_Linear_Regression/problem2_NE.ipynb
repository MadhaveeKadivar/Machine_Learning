# Simple Linear Regression using Normal Euqation

Importing dataset

```python
# Importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Image
import sklearn.metrics as sm
from sklearn import metrics
import math
from sklearn.metrics import accuracy_score
try:
    df = pd.read_csv('weatherHistory.csv')
except FileNotFoundError:
    print("File not found")
print(df.info())
```

Separating independent and dependent variables

```python
x = df['Humidity']
y = df['Apparent Temperature (C)']
```

Splitting Dataset for train and test

```python
s_f = 0.8
n = math.floor(s_f * df.shape[0])
l = len(x)
x_train = x.head(n)
y_train = y.head(n)
x_test = x.head(l-n)
y_test = y.head(l-n)

print("x_train")
print(x_train)
print("\ny_train")
print(y_train)
print("\nx_test")
print(x_test)
print("\ny_test")
print(y_test)
x_train.shape
```

Converting dataset into numpy array

```python
x_train = np.array(x_train)
print(x_train.shape)
y_train = np.array(y_train)
y_test = np.array(y_test)
y_train = y_train.reshape(-1,1)
y_test = y_test.reshape(-1,1)
```

Add one column in x train and x test array for x0 which is equal to one

```python
x0_train = np.ones((len(x_train),1))
X0_test = np.ones((len(x_test),1))
new_x_train = np.c_[x0_train,x_train]
new_x_test = np.c_[X0_test,x_test]
print(y_train.shape)
print(new_x_train.shape)
```

Implementing normal equation

```python
def theta(X,y):
    temp = np.dot(X.T,X)
    inverse = np.linalg.inv(temp)
    temp = np.dot(X.T,y)
    result = np.dot(inverse,temp)
    return result
```

Training data by calling function

```python
theta  =  theta(new_x_train,y_train)
print(theta)
```

Function for Calculating Hypothesis value of x train data 

```python
def hypothesis(x,theta):
    hypothesis_of_theta = np.dot(x,theta)
    return hypothesis_of_theta
```

# Testing

```python
y_predicted = hypothesis(new_x_test,theta)
```

```python
print(f"Accuracy of model : {metrics.r2_score(y_test,y_predicted)}")    
```
